## 3. Generate model

```python
from constants import *
print("YEAR_START:", YEAR_START)
print("YEAR_END:", YEAR_END)
print("COMPUTE_YEARS_BACKWARDS:", COMPUTE_YEARS_BACKWARDS)
print("YEARS_NO_TRAINING:", YEARS_NO_TRAINING)
print("CSV_FEATURES_LABEL:", CSV_FEATURES_LABEL)
```

```python
import pandas

df = pandas.read_csv(CSV_FEATURES_LABEL)
df
```

```python
df_clean = df.copy()
df_clean = df_clean[df_clean["year"] < (YEAR_END - YEARS_NO_TRAINING)]
df_clean.drop(['year', 'name'], axis=1, inplace=True)
df_clean.dropna(inplace=True)
df_clean
```

```python
ML_ALG_RANDOM_FOREST = 'random_forest'
ML_ALG_LOG_REGRESSION = 'log_regression'
ML_ALG_SUPPORT_VECTOR_MACHINES = 'svc'
ML_ALG_NEURAL_NETWORK = 'neural_network'

ML_ALG = ML_ALG_RANDOM_FOREST
```

```python
print("Generating model...")
print("ML algorithm:", ML_ALG)
print()

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

columns_features = list(df_clean.columns[0:-1])
column_label = df_clean.columns[-1]

X = df_clean[columns_features]
y = df_clean[column_label]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

if ML_ALG == ML_ALG_RANDOM_FOREST:
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier()
elif ML_ALG == ML_ALG_LOG_REGRESSION:
    from sklearn.linear_model import LogisticRegression
    model = LogisticRegression()
elif ML_ALG == ML_ALG_SUPPORT_VECTOR_MACHINES:
    from sklearn.svm import SVC
    model = SVC(kernel="linear")
elif ML_ALG == ML_ALG_NEURAL_NETWORK:
    from sklearn.neural_network import MLPClassifier
    from sklearn.model_selection import GridSearchCV
    clf = MLPClassifier(max_iter=100)
    parameter_space = {
        'hidden_layer_sizes': [(150,100,50), (10,30,10), (20,)],
        'activation': ['logistic', 'tanh', 'relu'],
        'solver': ['adam', 'lbfgs', 'sgd'],
        'alpha': [1e-5, 0.0001, 0.05],
        'learning_rate': ['constant', 'invscalin', 'adaptive'],
    }
    model = GridSearchCV(clf, parameter_space, n_jobs=-1, cv=5, verbose=2)
else:
    raise Exception("Unexpectred value of ML_ALG: %s." % ML_ALG)

model.fit(X_train, y_train)

if ML_ALG == ML_ALG_NEURAL_NETWORK:
    print('Best parameters found:', model.best_params_)

y_pred = model.predict(X_test)

print("Done.")

print()
print("Classification report:")
print(classification_report(y_test, y_pred))
```

```python
def predict_next(df, classifier):
    if len(set(df["year"])) > 1:
        raise Exception("Data from more than one year included in the dataset.")
    
    # remove year and name from beginning, and next_max from end
    X = df.copy()
    X = X.iloc[:, 2:-1]
    
    # predict
    pred = classifier.predict(X)
    
    # take only year and name
    df_pred = df.copy()
    df_pred = df_pred.iloc[:, 0:2]
    #add prediction
    df_pred["pred"] = pred
    
    # return names of possitives
    return list(df_pred[df_pred["pred"] > 0]["name"])
```

```python
print("Predictions of the years used for training and testing:")
print()

df_clean = df.dropna()

total_preds = 0
guessed_preds = 0

for year in range(YEAR_START, YEAR_END - YEARS_NO_TRAINING):
    print("%d " % (year + 1), end="")
    
    df_year = df_clean[df_clean["year"] == year]
    if len(df_year) == 0:
        print("Not enough data (minimum of %s years backwards needed)." % COMPUTE_YEARS_BACKWARDS)
        continue
    
    total_preds += 1
    
    next_max = df_year[df_year["next_max"] == 1].iloc[0]["name"]
    prediction = predict_next(df_year, model)
    
    if len(prediction) == 0:
        print("❌ ", end="")
        print("Predicted: %s. " % "none", end="")
    elif len(prediction) > 1:
        print("❌ ", end="")
        print("Predicted: %d (%s). " % (len(prediction), ", ".join(prediction)), end="")
    elif prediction[0] != next_max:
        print("❌ ", end="")
        print("Predicted: %s. " % prediction[0], end="")
    else:
        print("✅ ", end="")
        print("Predicted: %s. " % prediction[0], end="")
        guessed_preds += 1
    
    print("Most common name was: %s. " % next_max, end="")
    print()

print()
print("%d out of %d correct predictions (%.0f%%)." % \
    (guessed_preds, total_preds, guessed_preds * 100 / total_preds))
```
